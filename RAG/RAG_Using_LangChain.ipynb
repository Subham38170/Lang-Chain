```python
import os
```

```python
from youtube_transcript_api import YouTubeTranscriptApi,TranscriptsDisabled
from langchain_text_splitters import RecursiveCharacterTextSplitter
from langchain_huggingface import HuggingFaceEmbeddings,ChatHuggingFace,HuggingFaceEndpoint
from langchain_community.vectorstores import FAISS
from langchain_core.prompts import PromptTemplate
```

## Step 1a- Indexing (Document Ingeestion)

```python
video_id ='fGo-dv3p3hY'

try:
    yt_api = YouTubeTranscriptApi()
    transcript_list = yt_api.fetch(video_id,['en'])
    
except TranscriptsDisabled:
    print('No captions available for this video.')
```

```python
for chunk in transcript_list:
    print(chunk)
```

```python
transcript = ' '.join(chunk.text for chunk in transcript_list)
```

```python
transcript
```

## Step 1b : Indexing (Text Splitting)

```python
splitter = RecursiveCharacterTextSplitter(chunk_size=1000,chunk_overlap=200)
```

```python
chunks = splitter.create_documents([transcript])
```

```python
chunks
```

## Step 1c & 1d - Indexing (Embedding Generation and Storing in Vector Store)

```python
embedding = HuggingFaceEmbeddings(
    model_name = "sentence-transformers/all-MiniLM-L6-v2"
)
```

```python
vector_store = FAISS.from_documents(chunks,embedding)
```

```python
vector_store.index_to_docstore_id
```

```python
vector_store.get_by_ids(['9fd49aa8-b7dd-47f4-b370-f6e7aee12b8c'])
```

## Step 2 - Retrieval

```python
retriever = vector_store.as_retriever(
    search_type='similarity',
    search_kwargs={'k':4}
    )
```

```python
retriever.invoke('What is koin')
```

## Step 3 : Augmentation

```python
llm = HuggingFaceEndpoint(
    repo_id="openai/gpt-oss-20b", #google/gemma-2-2b-it
    task="conversational"
)

model = ChatHuggingFace(llm=llm)
```

```python
prompt = PromptTemplate(
    template='''
You are a helpful assistant,
Answer ONLY from the provided transcript context.
If context is insufficient, just say you don't know

{context}
Question : {question}
''',
input_variables=['context','question']
)
```

```python
question = 'What is koin dependency and how it is performed'
```

```python
retrieved_data = retriever.invoke(question)
```

```python
retrieved_data
```

```python
context_text= '\n\n'.join([doc.page_content for doc in retrieved_data])
```

```python
context_text
```

```python
final_prompt = prompt.invoke({'context':context_text,'question': question})
```

```python
final_prompt
```

## Step 4 : Generation

```python
answer = model.invoke(final_prompt)
```
