```python
import os
from dotenv import load_dotenv
load_dotenv(r'D:\AI ML\Gen AI\Lang_Chain\.env')
import openai
from IPython.display import Markdown

groq_api_key = os.getenv('GROQ_API_KEY')
```

```python
from langchain_groq import ChatGroq

model = ChatGroq(model='openai/gpt-oss-120b',api_key=groq_api_key)
```

```python
model
```

```python
from langchain_core.messages import HumanMessage,AIMessage,SystemMessage

messages = [
    SystemMessage('Tranlate the following from english to Hindi'),
    HumanMessage('Hello how are you')
]
Markdown(model.invoke(messages).content)
```

```python
from langchain_core.output_parsers import StrOutputParser
parser = StrOutputParser()
parser.invoke(model.invoke(messages).content)
```

```python
## Prompt Template
from langchain_core.prompts import ChatPromptTemplate

generic_template = 'Translate the following into {language}:'
prompt = ChatPromptTemplate([
    ('system',generic_template),
    ('user','{text}')
]
)
```

```python
result =prompt.invoke({
    'language': 'Hindi',
    'text' : 'Hello'
})
```

```python
result.to_messages()
```

```python
chain = prompt | model | parser
response = chain.invoke({
    'language' : 'Odia',
    'text': 'This is LLM'
})
Markdown(response)
```
