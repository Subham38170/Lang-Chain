```python
from langchain_huggingface import HuggingFaceEndpoint,ChatHuggingFace
import requests

llm = HuggingFaceEndpoint(
    repo_id='mistralai/Mixtral-8x7B-Instruct-v0.1', #google/gemma-2-2b-it
    task="text-generation"
)

model = ChatHuggingFace(llm=llm)
```

```python
from langchain_community.tools import DuckDuckGoSearchRun

search_tool = DuckDuckGoSearchRun()
results = search_tool.invoke('Top news in India today')
```

```python
from langchain.agents import create_agent
from langchainhub import Client
```

```python
#Step: 2 Pull the ReAct(Reasoning + Action) from Langchain Hub
client = Client()
prompt = client.pull('hwchase17/react')  #Pulls the standard ReAct agent prompt
prompt
```

```python
#Step3 : Create the ReAct agent manually with the pulled  prompt
agent = create_agent(
    model=model,
    tools=[search_tool],
    system_prompt=prompt
)
```

```python
response  = agent.invoke(
    {
        'input':'3 Ways to reach goa from delhi'
    }
)
```

```python
from langchain_community.tools import DuckDuckGoSearchRun
from langchain_openai import ChatOpenAI
from langgraph.prebuilt import create_react_agent

# Tool
search_tool = DuckDuckGoSearchRun()
tools = [search_tool]
agent = create_react_agent(model, tools)

# Run the query
response = agent.invoke({
    "input": "What is the capital of Odisha, its current population, "
             "and list tourist places with approximate travel + food + living costs."
})

# üü© Print final message (last one)
print("\nüü© Final Answer:\n")
if "messages" in response:
    final_message = response["messages"][-1].content
    print(final_message)
else:
    print("No messages found!")

# üß† Print all reasoning steps
print("\nüß† Agent Thought Process:\n")
for i, msg in enumerate(response["messages"]):
    role = type(msg).__name__
    print(f"Step {i+1} - {role}:")
    print(msg.content)
    print("-" * 60)
```
