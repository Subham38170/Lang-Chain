## Chroma

Chroma is a AI-native open- source vector database focused on devloper productivity and happiness. Chroma is licensed under Apache 2.0 .

```python
## Building a sample vectordb
from langchain_chroma import Chroma
from langchain_community.document_loaders import TextLoader
from langchain_community.embeddings import OllamaEmbeddings
from langchain_text_splitters import RecursiveCharacterTextSplitter
```

```python
loader = TextLoader('speech.txt')

docs = loader.load()
```

```python
docs
```

```python
text_splitter = RecursiveCharacterTextSplitter(chunk_size=500,chunk_overlap=0)
docs = text_splitter.split_documents(docs)
```

```python
docs
```

```python
embedding = OllamaEmbeddings(model='mxbai-embed-large')
vectordb = Chroma.from_documents(docs,embedding)
```

```python
vectordb
```

```python
query = 'What does the speaker believe is the main reason the United States should enter the war?'
docs =vectordb.similarity_search(query)
docs[0].page_content
```

```python
vector_db = Chroma.from_documents(docs,embedding,persist_directory='./Chroma_DB')
```

```python
db2 = Chroma(persist_directory='./Chroma_DB',embedding_function=embedding)
```

```python
db2.similarity_search(query)
```

```python
db2.similarity_search_with_score(query)
```

```python
retriver = vector_db.as_retriever()
retriver.invoke(query)
```
