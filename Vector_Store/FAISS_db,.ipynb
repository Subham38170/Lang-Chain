#### Facebool AI Similarity Search(FAISS) is a library for efficient similarity search and clustering of dense vectors. It contains algorithsm that search in sets of vectors of any size. up to ones that possibley do not fit in RAM. it also contains supporting code for evaluation and parameeter tuning.

```python
from langchain_community.document_loaders import TextLoader
from langchain_community.vectorstores import FAISS
from langchain_ollama.embeddings import OllamaEmbeddings
from langchain_text_splitters import CharacterTextSplitter
```

```python
loader = TextLoader('speech.txt')
documents = loader.load()
text_splitter = CharacterTextSplitter(chunk_size=200,chunk_overlap=30)
```

```python
docs = text_splitter.split_documents(documents)
```

```python
docs
```

```python
embedding = OllamaEmbeddings(model='mxbai-embed-large')
db = FAISS.from_documents(docs,embedding)
```

```python
query = 'What does the speaker describe the desired outcome of the war?'
```

```python
docs = db.similarity_search(query)
```

```python
docs
```

### As a Retriver

```python
retriever = db.as_retriever()
retriever.invoke(query)
```

### Similarity Search with score

There are some FAISS specific methods. One of them is similarity_search_with_score wheich allows you to return not only the docuemtns vbut also the distance score of the query to them, The returned distance score is L2 distance. Therefore, a lower score is better

```python
docs_and_score  = db.similarity_search_with_score(query)
```

```python
docs_and_score
```

```python
embedding_vector = embedding.embed_query(query)
```

```python
db.similarity_search_by_vector(embedding_vector)
```

```python
db.similarity_search_with_score_by_vector(embedding_vector)
```

```python
db.save_local('faiss_index')
```

```python
new_db =FAISS.load_local('faiss_index',embedding,allow_dangerous_deserialization=True)
```

```python
new_db.similarity_search(query)
```
