```python
from langchain_huggingface import HuggingFaceEndpoint,ChatHuggingFace
from dotenv import load_dotenv

```

```python
load_dotenv()
llm = HuggingFaceEndpoint(
    repo_id="Qwen/Qwen2.5-7B-Instruct", #google/gemma-2-2b-it
    task="conversational"

)

model = ChatHuggingFace(llm=llm)
```

```python
from langchain_core.tools import tool
from langchain_core.messages import HumanMessage
```

```python
@tool
def multiply(a: int,b: int) -> int:
    '''
    Given 2 numbers a and b this tool returns their product
    '''
    return a*b
```

```python
multiply.invoke({'a':4,'b':5})
```

```python
multiply.name,multiply.description
```

```python
multiply.args
```

```python
llm_with_tools = model.bind_tools([multiply])
```

```python
llm_with_tools.invoke('Hello how are you')
```

```python
llm_with_tools.invoke('Can you multiply 5 with 7')
```

```python
llm_with_tools.invoke('Can you multiply 5 with 7').tool_calls[0]
```

```python
result = llm_with_tools.invoke('Can you multiply 5 with 7')
```
