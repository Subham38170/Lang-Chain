```python
from langchain_text_splitters import HTMLHeaderTextSplitter
```

```python
html_code= '''
<!DOCTYPE html>
<html>
<head>
    <title>Small HTML Demo</title>
</head>
<body>

    <h1>My Small HTML Page</h1>

    <p>This is a simple page using some common <strong>HTML</strong> tags.</p>

    <h2>About</h2>
    <p>
        HTML means <abbr title="HyperText Markup Language">HTML</abbr>. 
        It is used to structure web pages.
    </p>

    <h2>List Example</h2>
    <ul>
        <li>First item</li>
        <li>Second item</li>
        <li>Third item</li>
    </ul>

    <h2>Image</h2>
    <img src="https://via.placeholder.com/150" alt="Sample Image">

    <h2>Contact Form</h2>
    <form>
        <label>Name:</label><br>
        <input type="text"><br><br>

        <label>Email:</label><br>
        <input type="email"><br><br>

        <button type="submit">Submit</button>
    </form>

</body>
</html>
'''
```

```python
headers_to_split_on = [
    ('h1','Header 1'),
    ('h2', 'Heaader 2'),
    ('h3','Header 3')
]

html_splitter = HTMLHeaderTextSplitter(headers_to_split_on)
```

```python
html_splitter.split_text(html_code)
```
